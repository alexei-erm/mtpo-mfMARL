```python
import pandas as pd
import matplotlib.pyplot as plt

df_mtmfq = pd.read_csv('multibattle_mtmfq.csv')
df_pomtmfq = pd.read_csv('multibattle_pomtmfq.csv')
```

```python
plt.plot(df_mtmfq['Episode'], df_mtmfq['Reward'].rolling(window=25).mean(), label='MTMFQ')
plt.plot(df_pomtmfq['Episode'], df_pomtmfq['Reward'].rolling(window=25).mean(), label='POMTMFQ')
plt.plot
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.legend()
plt.savefig('reward.png')
plt.show()
```
